```python
import matplotlib.pyplot as plt
valores = [('SP', 67.836,43), ('RJ', 36.678,66), ('MG', 29.229,88), ('ES', 27.165,48), ('Outros', 19.849,53)]
```

```python
nomes = [x[0] for x in valores]
valores = [x[1] for x in valores]

total = sum(valores)

porcentagens = [(x / total) * 100 for x in valores]

for i in range(0,5):
  print('Local:', nomes[i], '  Porcentagem: ',round(porcentagens[i], 2), '%')
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(porcentagens, labels=nomes, autopct='%1.2f%%')
ax1.axis('equal')
plt.show()
```
